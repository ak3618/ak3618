```python
#10 countries comparison of population growth comparison
#India vs Bangladesh, India vs China, India vs Pakistan, India vs Sri Lanka, India vs Nepal
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('countries.csv')
data
```

```python
#Requirment 1:coparison between India	Bangladesh
```

```python
India=data[data.country=='India']
India
```

```python
Bangladesh=data[data.country=='Bangladesh']
Bangladesh
```

```python
plt.plot(India.year,India.population/10**6)
plt.plot(Bangladesh.year,Bangladesh.population/10**6)
plt.xlabel('Year')
plt.ylabel('Population in billion')
plt.title('population growth comparison')
plt.legend(['India','Bangladesh'])
plt.show()
```
